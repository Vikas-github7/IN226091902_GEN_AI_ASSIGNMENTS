## Task 2: Multi-Input Prompt System

```python
# Define the multi-input template string
task2_template_str = "Explain {topic} for {audience} in a {tone} tone"

# Initialize the PromptTemplate
task2_prompt = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template=task2_template_str
)

# Define the requested test cases
test_cases_task2 = [
    {"topic": "AI", "audience": "beginners", "tone": "friendly"},
    {"topic": "Python", "audience": "kids", "tone": "fun"},
    {"topic": "Deep Learning", "audience": "engineers", "tone": "technical"}
]

# Loop through test cases and generate prompts dynamically
for case in test_cases_task2:
    result = task2_prompt.format(**case)
    print(f"Inputs: {case}\nResult: {result}\n")
```

## Task 4: ChatPromptTemplate System

```python
from langchain_core.prompts import ChatPromptTemplate

# Create a ChatPromptTemplate with System and User message roles
chat_template = ChatPromptTemplate.from_messages([
    ("system", "You are a professional {role}. Provide insights accordingly."),
    ("user", "Here is the topic to discuss: {topic}")
])

# Define the roles and topic to test
roles = ["teacher", "interviewer", "motivator"]
chat_topic = "Neural Networks"

for role in roles:
    # Format messages for each role
    messages = chat_template.format_messages(role=role, topic=chat_topic)
    
    print(f"Role: {role.capitalize()}")
    for msg in messages:
         print(f" - {msg.type.capitalize()}: {msg.content}")
    print("-" * 20)
```

## Task 5: Input Validation Layer 

```python
def validate_inputs(audience, tone):
    """
    Validates audience and tone parameters. Assigns default values if invalid.
    """
    valid_audiences = ["beginner", "intermediate", "expert"]
    valid_tones = ["formal", "casual", "fun"]
    
    # Assign defaults if inputs are outside the allowed lists
    final_audience = audience if audience in valid_audiences else "beginner"
    final_tone = tone if tone in valid_tones else "casual"
    
    return final_audience, final_tone

print("Valid inputs ('expert', 'formal') ->", validate_inputs("expert", "formal"))
print("Invalid inputs ('kids', 'angry') ->", validate_inputs("kids", "angry"))
```

## Task 6: Prompt Generator App

```python
def generate_prompt(topic, audience, tone, style):
    """
    Validates inputs and builds a prompt string dynamically based on parameters.
    """
    # 1. Pass inputs through validation layer
    val_audience, val_tone = validate_inputs(audience, tone)
    
    # 2. We use a base PromptTemplate layout tailored to the requested example output
    app_template_str = "Explain {topic} for {audience}s in a {tone} {style} style"
    
    app_prompt = PromptTemplate(
        input_variables=["topic", "audience", "tone", "style"],
        template=app_template_str
    )
    
    # 3. Format and return
    return app_prompt.format(
        topic=topic, 
        audience=val_audience, 
        tone=val_tone, 
        style=style
    )


# Example matching the assignment requirement:
# Expects: Explain Neural Networks for beginners in a fun storytelling style
generated_result = generate_prompt("Neural Networks", "beginner", "fun", "storytelling")
print(generated_result)
```

## Task 7: Template Reusability Test

```python
# 1. Define ONE reusable template
reusable_template = PromptTemplate(
    input_variables=["action", "subject"],
    template="Please {action} the following concept: {subject}"
)

# 2. Define 5 different input sets
test_inputs = [
    {"action": "summarize", "subject": "Generative AI"},
    {"action": "write a poem about", "subject": "Python programming"},
    {"action": "create a quiz for", "subject": "Data Structures"},
    {"action": "translate to Spanish", "subject": "Machine Learning"},
    {"action": "find the flaws in", "subject": "Linear Regression"}
]

# 3. Run the template with all 5 sets
for i, inputs in enumerate(test_inputs, 1):
    print(f"Run {i}: {reusable_template.format(**inputs)}")
```
